```python
import json
import pandas
```

```python
json_data_baskar=open('statsDict_Baskar.json')
json_data_john=open('statsDict_john_lee.json')
json_data_jeongwoo=open('statsDict_jeongwoo_chang.json')
json_data_kyle=open('statsDict_kyle_mckiou.json')
json_data_mylene=open('statsDict_mylene.json')
json_data_peyman=open('statsDict_Peyman_Mahdian.json')
json_data_sanorita=open('statsDict_sanorita.json')
json_data_victor=open('statsDict_Victor_Ge.json')
json_data_ashutosh=open('statsDict_Ashutosh_Dhekne.json')

data_baskar = json.load(json_data_baskar)
data_john = json.load(json_data_john)
data_jeongwoo = json.load(json_data_jeongwoo)
data_kyle = json.load(json_data_kyle)
data_mylene = json.load(json_data_mylene)
data_peyman = json.load(json_data_peyman)
data_sanorita = json.load(json_data_sanorita)
data_victor = json.load(json_data_victor)
data_ashutosh = json.load(json_data_ashutosh)


json_data_baskar.close()
json_data_john.close()
json_data_jeongwoo.close()
json_data_kyle.close()
json_data_mylene.close()
json_data_peyman.close()
json_data_sanorita.close()
json_data_victor.close()
json_data_ashutosh.close()
```

```python
df1 = pandas.DataFrame.from_dict(data_baskar,orient='index')#, dtype=None)
df2 = pandas.DataFrame.from_dict(data_john,orient='index')#, dtype=None)
df3 = pandas.DataFrame.from_dict(data_jeongwoo,orient='index')#, dtype=None)
df4 = pandas.DataFrame.from_dict(data_kyle,orient='index')#, dtype=None)
df5 = pandas.DataFrame.from_dict(data_mylene,orient='index')#, dtype=None)
df6 = pandas.DataFrame.from_dict(data_peyman,orient='index')#, dtype=None)
df7 = pandas.DataFrame.from_dict(data_sanorita,orient='index')#, dtype=None)
df8 = pandas.DataFrame.from_dict(data_victor,orient='index')#, dtype=None)
df9 = pandas.DataFrame.from_dict(data_ashutosh,orient='index')#, dtype=None)
```
